### 1. Cálculo del Índice de Madurez

Este código ayuda a determinar si una fruta está lista para cosecha basado en la relación entre sólidos solubles (°Brix) y acidez.

```
# Datos de la muestra
brix = 12.5
acidez = 0.8

# Cálculo del índice
indice_madurez = brix / acidez

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Fruta lista para cosecha/consumo.")
else:
    print("Estado: Aún le falta madurar.")
```

### 2. Conversión de Unidades de Temperatura (Pasteurización)

Muchos equipos industriales vienen en Fahrenheit, pero nosotros trabajamos en Celsius. Este pequeño script automatiza la conversión para un proceso de control.

```
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

temp_pasteurizacion = 72 # Grados Celsius
temp_f = celsius_a_fahrenheit(temp_pasteurizacion)

print(f"Para pasteurizar a {temp_pasteurizacion}°C, ajusta el equipo a {temp_f}°F.")
```

### 3. Estimación de Rendimiento de Extracción

Ideal para calcular cuántos litros de aceite o jugo obtienes por cada kilo de materia prima.

```
peso_fruta_kg = 500  # Materia prima que entra
peso_jugo_kg = 220   # Producto obtenido

rendimiento = (peso_jugo_kg / peso_fruta_kg) * 100

print(f"El rendimiento de la extracción fue del {rendimiento}%")
print(f"Se perdieron {peso_fruta_kg - peso_jugo_kg} kg en bagazo/residuos.")
```
